```python

# Concatenate Firstname and Surname
reportsDF['Fullname'] = reportsDF['First name'] + ' ' + reportsDF['Surname']

# Print the result
print(reportsDF[['Fullname', 'ID number', 'State', 'Started on', 'Completed', 'Time taken', 'Grade/10.00', 'Q. 1 /10.00', 'questionID']])
```

```python
cols = reportsDF.columns.tolist()
cols.remove('ID number')
cols.append('ID number')
reportsDF = reportsDF[cols]
reportsDF.info()
```

```python
import hashlib

# Hàm để hash giá trị trong cột
def hash_column_value(value):
    # Sử dụng SHA-256 để hash giá trị
    hashed_value = hashlib.sha256(value.encode()).hexdigest()
    return hashed_value

# Áp dụng hàm hash_column_value cho cột "Fullname"
reportsDF['Fullname'] = reportsDF['Fullname'].apply(hash_column_value)

# Áp dụng hàm hash_column_value cho cột "ID number"
reportsDF['ID number'] = reportsDF['ID number'].apply(hash_column_value)
```

```python
# Xuất DataFrame ra file CSV
reportsDF.to_csv('reports.csv', index=False)
```
